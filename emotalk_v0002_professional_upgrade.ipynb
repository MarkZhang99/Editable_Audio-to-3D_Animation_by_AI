## train emotalk_model_v0002_upgrade

```python
# updated_train.py
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader
from emotalk_model_v0002_upgrade import EmoTalkUpgrade
from train import DummyDataset
from loss_v0002 import EmoTalkLoss
from types import SimpleNamespace
from tqdm import tqdm
import matplotlib.pyplot as plt
import os

os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

# === Configuration ===
DEVICE      = 'cuda' if torch.cuda.is_available() else 'cpu'
EPOCHS      = 10
LR          = 1e-4
SAVE_PATH   = 'emotalk_upgrade_finetuned.pth'
BATCH_SIZE  = 1
CLS_ALPHA   = 0.1   # weight for classification loss

args = SimpleNamespace(
    bs_dim            = 52,
    device            = DEVICE,
    max_seq_len       = 512,
    period            = 20,
    emo_gru_hidden    = 128,
    emo_gru_layers    = 2,
    transformer_layers= 4,
    transformer_heads = 8,
    transformer_dim   = 512,
    num_emotions      = 8,
    num_person        = 25,
    dropout           = 0.1,
    use_prosody       = False,
)

# === Data loaders ===
train_loader = DataLoader(DummyDataset('./data/RAVDESS/train'),
                          batch_size=BATCH_SIZE, shuffle=True)
val_loader   = DataLoader(DummyDataset('./data/RAVDESS/val'),
                          batch_size=BATCH_SIZE, shuffle=False)
test_loader  = DataLoader(DummyDataset('./data/RAVDESS/test'),
                          batch_size=BATCH_SIZE, shuffle=False)

# === Model, Loss, Optimizer ===
model    = EmoTalkUpgrade(args).to(DEVICE)
loss_fn  = EmoTalkLoss(region_weighted=True)           
optimizer= torch.optim.Adam(model.parameters(), lr=LR)

train_loss_list, val_loss_list = [], []
val_cls_list = []

# === Training loop ===
for epoch in range(1, EPOCHS+1):
    model.train()
    total_bs_loss  = 0.0
    total_cls_loss = 0.0
    logs_accum     = {"main":0, "smooth":0, "vel":0, "total":0}

    for batch in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS} [Train]"):
        audio      = batch["audio"].to(DEVICE)
        blendshape = batch["blendshape"].to(DEVICE)
        level      = batch["level"].to(DEVICE)
        person     = batch["person"].to(DEVICE)
        args.target_len = blendshape.size(1)

        # forward
        bs_preds, cls_logits = model(audio, level, person)
        bs_loss, logs        = loss_fn(bs_preds, blendshape)
        cls_loss             = F.cross_entropy(cls_logits, level)
        loss                 = bs_loss + CLS_ALPHA * cls_loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_bs_loss  += bs_loss.item()
        total_cls_loss += cls_loss.item()
        for k,v in logs.items():
            logs_accum[k] += v

    # compute averages
    avg_bs  = total_bs_loss  / len(train_loader)
    avg_cls = total_cls_loss / len(train_loader)
    avg_trn = avg_bs + CLS_ALPHA * avg_cls
    train_loss_list.append(avg_trn)

    print(f"[Epoch {epoch}] BS Loss: {avg_bs:.4f} | CLS Loss: {avg_cls:.4f}")
    print(f"  -> main:{logs_accum['main']:.4f} smooth:{logs_accum['smooth']:.4f} "
          f"vel:{logs_accum['vel']:.4f} total:{logs_accum['total']:.4f}")

    # === Validation ===
    model.eval()
    val_bs = 0.0
    # val_cls = 0.0  # if you want to track classification on val
    with torch.no_grad():
        for batch in tqdm(val_loader, desc=f"Epoch {epoch}/{EPOCHS} [Val]"):
            audio      = batch["audio"].to(DEVICE)
            blendshape = batch["blendshape"].to(DEVICE)
            level      = batch["level"].to(DEVICE)
            person     = batch["person"].to(DEVICE)
            args.target_len = blendshape.size(1)

            bs_preds, cls_logits = model(audio, level, person)
            loss_bs, _          = loss_fn(bs_preds, blendshape)
            val_bs += loss_bs.item()
            # val_cls += F.cross_entropy(cls_logits, level).item()

    avg_val_bs = val_bs / len(val_loader)
    val_loss_list.append(avg_val_bs)
    # val_cls_list.append(val_cls / len(val_loader))

    print(f"  -> Val BS Loss: {avg_val_bs:.4f}")
    # print(f"  -> Val CLS Loss: {val_cls_list[-1]:.4f}")

    torch.save(model.state_dict(), SAVE_PATH)

print("✅ Training complete — model saved to", SAVE_PATH)

# === Test evaluation ===
print("\n=== Final Test ===")
model.load_state_dict(torch.load(SAVE_PATH))
model.eval()

test_bs_accum = 0.0
correct = total = 0
with torch.no_grad():
    for batch in tqdm(test_loader, desc="Test"):
        audio      = batch["audio"].to(DEVICE)
        blendshape = batch["blendshape"].to(DEVICE)
        level      = batch["level"].to(DEVICE)
        person     = batch["person"].to(DEVICE)
        args.target_len = blendshape.size(1)

        bs_out, cls_logits = model(audio, level, person)
        loss_bs, _        = loss_fn(bs_out, blendshape)
        test_bs_accum    += loss_bs.item()

        preds   = cls_logits.argmax(dim=1)
        correct += (preds == level).sum().item()
        total   += level.size(0)

test_bs = test_bs_accum / len(test_loader)
acc     = correct / total
print(f"📊 Test BS Loss: {test_bs:.4f} | Test Acc: {acc:.4f}")

# === Plot ===
plt.figure(figsize=(8,5))
plt.plot(train_loss_list, label='Train (BS+CLS)')
plt.plot(val_loss_list,   label='Val (BS)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training & Validation Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('loss_curve.png')
plt.show()
```

Analysis of First-Round Experiments (A–D)\
In the initial grid search we compared four settings:\
A (LR=1e-5, CLS_ALPHA=0.1): Fastest convergence, lowest train BS loss (0.0333) and strong generalization (val BS 0.0232).\
B (LR=3e-6, CLS_ALPHA=0.1): Slightly higher train loss (0.0419) but best val BS loss (0.0229), indicating better overfit control.\
C (LR=1e-6, CLS_ALPHA=0.1): LR too small—slow, under-fitting (val BS 0.0265).\
D (LR=1e-6, CLS_ALPHA=0.2): Boosting classification weight only marginally improved val BS (0.0265→0.0258) but did not recover from the overly low LR.

Key takeaways:\
Optimal LR lies between 3 × 10⁻⁶ and 1 × 10⁻⁵.\
A moderate classification weight (CLS_ALPHA≈0.1) is sufficient; larger values yield diminishing returns on the regression task.
